## Motivation

Deep learning is a rich field of machine learning that currently enjoys a large amount of academic research as well as commerical success. One specific architecture of deep learning that has been used for a wide array of image categorization problems is convolutional neural networks (CNN). CNNs are well equipped for problems involving image data as the neurons (including hidden layers) are arranged in 3 dimensions (height, width, depth) and thus fit the general structure of the image data$^{1,2}$.

## Project Description

The goal of this project is to train a CNN model that can recognize and categorize facial expressions of facial images based on training image data. The training data set comes from Kaggle$^3$ which has a facial image training set consisting of 28,709 examples and a facial image testing set consisting of 3,589 examples. The goal of the project is to train a model to categorize each face based on its emotion into one of seven categories (0=Angry, 1=Disgust, 2=Fear, 3=Happy, 4=Sad, 5=Surprise, 6=Neutral).

## Algorithm

In short, CNNs take an image, move the image data through a series of layers (convolutional, ReLU (non-linear), pooling, and fully connected layers) in order to get an output. In our particular case, the output will be a single class (facial expression category). 

```python
from IPython.display import Image
Image(url="https://adeshpande3.github.io/assets/Table.png", embed=True)
```

**Convolutional Layers**

The convolutional layer uses a filter over different regions (i.e., receptive fields) of the image until it covers the entire image. As the filter slides over the different receptive fields, the weights embedded in the filter are multiplied element-wise with the pixel values and summed. This operation is repeated for every possible receptive field that covers the entire image.

```python
from IPython.display import Image
Image(url="https://adeshpande3.github.io/assets/ActivationMap.png", embed=True)
```

The convultional layer effectively helps identify features of the image. For example, a specific filter could be used to detect curves in the image. 

```python
from IPython.display import Image
Image(url="https://adeshpande3.github.io/assets/Filter.png", embed=True)
```

If this filter is applied to a receptive field that does indeed have a curve, the element-wise multiplication will be large; otherwise, it will be small.

```python
from IPython.display import Image
Image(url="https://adeshpande3.github.io/assets/SecondMultiplication.png", embed=True)
```

**ReLU Layers**

ReLU (Rectified Linear Units) layers are filters that apply the non-saturating activation function $f(x)=\max(0,x)^1$. These help apply non-linear properties while preserving the dimensions of the neural network.

**Pool**

The pooling layers provide "down-sampling" which means that it looks at the non-linear features from previous layers and attemps to integrate them together into a single layer. These are integrated together in such a way that the maximum non-linear features are stitched together in a non-overlapping way.

```python
from IPython.display import Image
Image(url="https://upload.wikimedia.org/wikipedia/commons/e/e9/Max_pooling.png", embed=True)
```

**Fully Connected Layer**

After all of the convolutional and ReLU layers are applied to identify different types of features and non-linearities, the fully connected layer effectively takes the final vector of outcomes (expressed as probabilities). These probabilities are found by looking at the activation maps from previous filters and finding which features the previous high level filters correlate to.
